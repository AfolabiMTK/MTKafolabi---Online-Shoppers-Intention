## CASE STUDY

The data used in this analysis is an Online Shoppers Purchasing Intention data set provided from a top ecommerce site. The data set was formed so that each session would belong to a different user in a 1-year period to avoid any tendency to a specific campaign, special day, user profile, or period.

The primary purpose of the data set is to predict the purchasing intentions of a visitor to this particular storeâ€™s website. This dataset has very few missing values and all features of the dataset are relevant to the purchasing intention based on inference.


### FEATURE DESCRIPTION

**The dataset consists of 18 features (columns) with 12330 rows.**

**The features of the dataset is described below:**

| COLUMNS | DESCRIPTION |
| --- | --- |
| Administrative: | This is the number of pages of this type (administrative) that the user visited. |
| Administrative_Duration: | This is the amount of time spent in this category of pages. |
| Informational: | This is the number of pages of this type (informational) that the user visited. |
| Informational_Duration: | This is the amount of time spent in this category of pages. |
| ProductRelated: | This is the number of pages of this type (product related) that the user visited. |
| ProductRelated_Duration: | This is the amount of time spent in this category of pages. |
| BounceRates: | The percentage of visitors who enter the website through that page and exit without triggering any additional tasks. |
| ExitRates: | The percentage of pageviews on the website that end at that specific page. |
| PageValues: | The average value of the page averaged over the value of the target page and/or the completion of an eCommerce transaction. |
| SpecialDay: | This value represents the closeness of the browsing date to special days or holidays (eg Mother's Day or Valentine's day) in which the transaction is more likely to be finalized. More information about how this value is calculated below. |
| Month: | Contains the month the pageview occurred, in string form. |
| OperatingSystems: | An integer value representing the operating system that the user was on when viewing the page. |
| Browser: | An integer value representing the browser that the user was using to view the page. |
| Region: | An integer value representing which region the user is located in. |
| TrafficType: | An integer value representing what type of traffic the user is categorized into. |
| VisitorType: | A string representing whether a visitor is New Visitor, Returning Visitor, or Other. |
| Weekend: | A boolean representing whether the session is on a weekend. |
| Revenue: | A boolean representing whether or not the user completed the purchase. |

```python
# Page Metrics Analysis

fig = plt.figure(figsize=(16, 4))

ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

sns.distplot(osi['BounceRates'], bins=20, ax=ax1)
sns.distplot(osi['ExitRates'], bins=20, ax=ax2)
sns.distplot(osi['PageValues'], bins=20, ax=ax3)

plt.tight_layout()
plt.show()
```

It should be noted that:

- BounceRates is the percentage where the first page visited was the only page visited in that session. 
- ExitRates is the percentage where the page was the last page visited in the session, out of all visits. 

Then the visualizations above shows three google analytics metrics which can be said to be that:

- BounceRates & ExitRates follow a normal distribution unlike PageValues.
- All the three distributions contains outliers.
- All the three features have distributions that are skewed right. 
- The ExitRates has more high values than BounceRates, because, it is very possible that the transaction confirmation pages for will cause the average exit rate to increase.

```python
# Visit Date Analysis

fig = plt.figure(figsize=(18, 12))

ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

orderlist = ['Jan','Feb','Mar','Apr','May','June','Jul','Aug','Sep','Oct','Nov','Dec']
sns.countplot(data=osi, x='Month', hue='Revenue', ax=ax1, order=orderlist)
sns.countplot(data=osi, x='SpecialDay', hue='Revenue', ax=ax2)

plt.tight_layout()
plt.show()

fig, ax = plt.subplots(1, 2,figsize=(12, 6), subplot_kw=dict(aspect="equal"))

ax[0].pie(osi['Weekend'].value_counts(),explode=(0.1,0),labels=['Weekday','Weekend'], autopct='%1.0f%%')
ax[0].set_title('Weekend vs. Weekday (Total Visits)')

ax[1].pie(osi[osi['Revenue'] == 1]['Weekend'].value_counts(),explode=(0.1,0),labels=['Weekday','Weekend'], autopct='%1.0f%%')
ax[1].set_title('Weekend vs. Weekday (Only Visits Ended with Transactions)')

plt.title('WEEKEND VISITIS', fontsize = 30)

plt.show()
```

With the visit date analysis done above, I could deduce that;

- 1. March, May, November & December have a lot of visits. May is the month with the highest number of visits, yet transactions made is not up to November. 
- 2. There was no visits at all during January & April.
- 3. Most transactions happen during the end of the year, with Nov as the month with the highest number of confirmed transactions. 
- 4. Most of transactions happen on special days (SpecialDay = 0). 
- 5. There was increase in the number of transactions happening on weekends compared to those on weekdays. 

Meanwhile, any visit date to a special day like Christians or Muslims holidays, black Friday, new year, and so on, there is assuarnce that it will end up in a transaction because people will want to take it as an opprotunity.
